```python
import pydicom
import matplotlib.pyplot as plt

# Ruta al archivo DICOM
file_path = 'C:/Users/lcres/PycharmProjects/Flask Server Brain Tumor/GLIOMA/1.2.840.113845.11.1000000001951524609.20160401120008.1117382'

# Leer el archivo DICOM
dicom_data = pydicom.dcmread(file_path)

# Extraer la imagen
image = dicom_data.pixel_array

# Visualizar la imagen
plt.imshow(image, cmap='gray')
plt.title('Imagen DICOM')
plt.axis('off')  # Opcional: quita los ejes para una mejor visualización
plt.show()
```

```python
import pydicom

# Ruta al archivo DICOM


# Leer el archivo DICOM
dicom_data = pydicom.dcmread(file_path)

# Verificar el número de imágenes
if hasattr(dicom_data, 'NumberOfFrames'):
    print(f"El archivo contiene {dicom_data.NumberOfFrames} imágenes.")
    # Para visualizar cada frame si es necesario
    for i in range(dicom_data.NumberOfFrames):
        plt.imshow(dicom_data.pixel_array[i], cmap='gray')
        plt.title(f'Imagen {i+1}')
        plt.axis('off')
        plt.show()
else:
    print("El archivo contiene una sola imagen.")
    plt.imshow(dicom_data.pixel_array, cmap='gray')
    plt.title('Imagen DICOM')
    plt.axis('off')
    plt.show()
```
